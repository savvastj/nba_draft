```python
import pandas as pd
```

```python
trad_df = pd.read_csv("raw_data/cbb_player_season_trad_stats_07_08_17_raw.csv")
adv_df = pd.read_csv("raw_data/cbb_player_season_adv_stats_07_08_17_raw.csv")
```

```python
trad_df.info()
```

```python
adv_df.head()
```

```python
adv_players = adv_df.Player.str.split(', ', expand=True)
```

```python
adv_players[0] = adv_players[0].str.strip("(\'")
```

```python
adv_players[1] = adv_players[1].str.strip(")\'").str.extract("/.*/.*/(.*)\.", 
                                                             expand=False)
```

```python
adv_players.head()
```

```python
adv_players.values[:2]
```

```python
adv_df[["Player", "Sref_Cbb_ID"]] = adv_players[[0, 1]]
```

```python
adv_df.head()
```

```python
trad_players = trad_df.Player.str.split(', ', expand=True)
trad_players[0] = trad_players[0].str.strip("(\'")
trad_players[1] = trad_players[1].str.strip(")\'").str.extract("/.*/.*/(.*)\.", 
                                                             expand=False)
trad_players.head()
```

```python
trad_df[["Player", "Sref_Cbb_ID"]] = trad_players[[0, 1]]
```

```python
trad_df.head()
```

```python
error_trad_df = pd.read_csv("raw_data/error_df_trad_stats_07_09_17.csv")
```

```python
error_trad_df.head()
```

```python
error_trad_df.info()
```

```python
trad_df.info()
```

```python
trad_df2 = pd.concat([trad_df, error_trad_df], ignore_index=True)
```

```python
trad_df2.info()
```

```python
adv_df.info()
```

## Clean up column names

```python
adv_df.columns = adv_df.columns.str.replace("%", "_Pct")
```

```python
trad_df2.columns.values[11:15] = ["FG_" + col for col in 
                                  trad_df2.columns.values[11:15]]
```

```python
adv_df.columns
```

```python
trad_df2.columns
```

## Create some FG% columns

```python
trad_df2["FG_Pct"] = trad_df2.FG / trad_df2.FGA
trad_df2["FG_2P_Pct"] = trad_df2.FG_2P / trad_df2.FG_2PA
trad_df2["FG_3P_Pct"] = trad_df2.FG_3P / trad_df2.FG_3PA
```

```python
trad_df2.head()
```

```python
# almost forgot FT%
trad_df2["FT_Pct"] = trad_df2.FT / trad_df2.FTA
```

```python
trad_df2.head()
```

```python
trad_df2.head()
```

## Drop Rk column

```python
trad_df2.drop("Rk", axis=1, inplace=True)
adv_df.drop("Rk", axis=1, inplace=True)
```

```python
trad_df2.info()
```

```python
adv_df.info()
```

```python
adv_df.to_csv("processed_data/cbb_player_seasons_adv_stats_07_09_17.csv",
              index=False)
trad_df2.to_csv("processed_data/cbb_player_season_trad_stats_07_09_17.csv",
                index=False)
```
